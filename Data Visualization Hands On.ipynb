```python
import pandas as pd
df = pd.read_csv('https://query.data.world/s/lx5gnlyrog5hozt6sewsw2mqh4jq4s')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
plt.hist(df.MonthlyCharges)
```

```python
sns.distplot(df.MonthlyCharges, hist=False)
```

```python
Churn = df[df['Churn']=='Yes']
```

```python
NoChurn = df[df['Churn']=='No']
```

```python
sns.distplot(Churn.MonthlyCharges, hist=False)
```

```python
sns.distplot(NoChurn.MonthlyCharges, hist=False)
```

```python
sns.distplot(Churn.MonthlyCharges, hist=False, label = 'Churn')
sns.distplot(NoChurn.MonthlyCharges, hist=False, label = 'Non Churn')
plt.legend()
```

```python
sns.boxplot(Churn.MonthlyCharges)
```

```python
sns.boxplot(NoChurn.MonthlyCharges)
```

```python
sns.boxplot(Churn.tenure)
```

```python
sns.boxplot(NoChurn.tenure)
```

```python
plt.figure(figsize = (20,5))#Pixels: width, height
plt.subplot(1,2,1)
plt.title("Churn Tenure")
sns.boxplot(Churn.tenure)

plt.subplot(1,2,2)
plt.title("Non-Churn Tenure")
sns.boxplot(NoChurn.tenure)
```

```python
sns.boxplot(x = df['Churn'], y = df['tenure'])
```

```python
sns.countplot(df['Churn'])
```

```python
plt.figure(figsize = (20,5))#Pixels: width, height
plt.subplot(1,2,1)
plt.title("Churn Contract")
sns.countplot(Churn['Contract'])
plt.subplot(1,2,2)
plt.title("No Churn Contract")
sns.countplot(NoChurn['Contract'])
```
